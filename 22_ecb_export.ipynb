# ECB Export from excel to graphic

```python
import pandas as pd
from matplotlib import pyplot as plt
from datetime import date

file_name = 'data/ECB Data Portal wide_20241120094402.xlsx'
columns = ['DATE', 'HICP - Overall index" (ICP.M.PT.N.000000.4.ANR)', 'HICP - Overall index (ICP.M.U2.N.000000.4.ANR)']

df = pd.read_excel(file_name, sheet_name='DATA', usecols=columns, index_col='DATE', parse_dates=['DATE'], skiprows=14)
df.columns=['PT', 'EU']

df.to_excel('inflation.xlsx')
df.plot(title='Inflation Ratio', legend=True, figsize=(10,4), grid=True, ylabel='%')

plt.savefig(f'inflation-rate-{date.today()}.png')
```
