```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler


np.random.seed(42)
dates = pd.date_range(start='2000-01-01', periods=100, freq='Q')

data = {
    'PIB': np.random.normal(2, 0.5, 100),
    'Inflacao': np.random.normal(2, 0.8, 100),
    'Desemprego': np.random.normal(5, 1.0, 100),
    'Taxa_Juros': np.random.normal(3, 0.7, 100),
    'Prod_Industrial': np.random.normal(1.5, 0.4, 100)
}

df_macro = pd.DataFrame(data, index=dates)


scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_macro)

pca = PCA(n_components=2)
pca_result = pca.fit(df_scaled)

eigenvalues = pca.explained_variance_
explained_variance_ratio = pca.explained_variance_ratio_
eigenvectors = pca.components_

# Normalizando os valores próprios de -1 a 1
eigenvectors_normalized = eigenvectors / np.max(np.abs(eigenvectors), axis=1, keepdims=True)

print("Valores próprios (Eigenvalues):")
print(eigenvalues)

variances = []
print("\nRazão de Variância Explicada (%):")
for i, var in enumerate(explained_variance_ratio):
    print(f"Componente Principal {i + 1}: {var * 100:.2f}%")
    variances.append(var * 100)

variances_ratio = sum(variances)

print("\nValores próprios (Eigenvectors) normalizados:")

for i, eigenvector in enumerate(eigenvectors_normalized):
    print(f"Componente Principal {i + 1}: {eigenvector}")

df_eigenvectors = pd.DataFrame(eigenvectors_normalized, columns=df_macro.columns)
df_eigenvectors = df_eigenvectors.T.reset_index()
df_eigenvectors.columns = ['index', 'PC1', 'PC2']

print("\nDataFrame dos Valores próprios normalizados:")
print(df_eigenvectors)
```
