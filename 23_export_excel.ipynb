```python
from sklearn import datasets

wine = datasets.load_wine()
print(wine.DESCR)
```

```python
import pandas as pd

df = pd.DataFrame(wine.data, columns=wine.feature_names)
df
```

```python
file = pd.ExcelWriter('wine_report.xlsx', engine='xlsxwriter')
df.to_excel(file, sheet_name='Sheet1', startrow=2)
book = file.book
sheet = file.sheets['Sheet1']
font_bold = book.add_format({'bold': True, 'size': 24})
sheet.write('A1', 'Wine report', font_bold)

# chart
chart = book.add_chart({'type': 'column'})
chart.add_series({'values': '=Sheet1!B4:B8', 'name': '=Sheet1!B3'})
sheet.insert_chart('P4', chart)
file.close()
```

```python
from ipywidgets import interact_manual, IntRangeSlider

def limit_dataframe(range:list):
    """ Limits the output of dataframe """

    df2 = df[range[0]:range[1]+1]  # change
    file = pd.ExcelWriter('wine_report.xlsx', engine='xlsxwriter')
    df2.to_excel(file, sheet_name='Sheet1', startrow=2)
    book = file.book
    sheet = file.sheets['Sheet1']
    font_bold = book.add_format({'bold': True, 'size': 24})
    sheet.write('A1', 'Wine report', font_bold)

    # chart
    chart = book.add_chart({'type': 'column'})
    chart.add_series({'values': '=Sheet1!B4:B8', 'name': '=Sheet1!B3'})
    sheet.insert_chart('P4', chart)
    file.close()
    print('File saved')
    return df2

interact_manual(limit_dataframe, range=IntRangeSlider(max=len(df)));
```
